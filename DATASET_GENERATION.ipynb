# Complete Dataset Generator

```python
import pandas as pd
import numpy as np
import random
import os
from sklearn.preprocessing import MinMaxScaler

# Reproducibility
random.seed(42)
np.random.seed(42)

# Number of players
n_players = 3000

roles = ["Batter", "Bowler", "All-Rounder", "Wicketkeeper"]

data = []

for i in range(n_players):

    # Basic Information
    player_id = i + 1
    age = random.randint(18, 35)
    role = random.choice(roles)
    matches = random.randint(10, 250)

    # -------------------------
    # Role-Based Statistics
    # -------------------------

    if role == "Batter":

        runs = random.randint(3000, 8000)
        batting_average = round(random.uniform(35, 65), 2)
        strike_rate = round(random.uniform(120, 190), 2)

        fifties = random.randint(10, 50)
        hundreds = random.randint(1, 20)

        wickets = random.randint(0, 30)
        economy = round(random.uniform(6, 10), 2)

        catches = random.randint(10, 80)

    elif role == "Bowler":

        runs = random.randint(0, 1500)
        batting_average = round(random.uniform(10, 30), 2)
        strike_rate = round(random.uniform(70, 120), 2)

        fifties = random.randint(0, 5)
        hundreds = random.randint(0, 1)

        wickets = random.randint(100, 350)
        economy = round(random.uniform(4, 8), 2)

        catches = random.randint(5, 50)

    elif role == "All-Rounder":

        runs = random.randint(1500, 5000)
        batting_average = round(random.uniform(25, 50), 2)
        strike_rate = round(random.uniform(100, 160), 2)

        fifties = random.randint(5, 30)
        hundreds = random.randint(0, 10)

        wickets = random.randint(50, 250)
        economy = round(random.uniform(4, 8), 2)

        catches = random.randint(10, 100)

    else:  # Wicketkeeper

        runs = random.randint(2000, 7000)
        batting_average = round(random.uniform(25, 55), 2)
        strike_rate = round(random.uniform(100, 170), 2)

        fifties = random.randint(5, 40)
        hundreds = random.randint(0, 15)

        wickets = random.randint(0, 10)
        economy = round(random.uniform(6, 10), 2)

        catches = random.randint(50, 150)

    # -------------------------
    # Age-Based Fitness
    # -------------------------

    if age <= 24:
        fitness_score = random.randint(80, 100)

    elif age <= 30:
        fitness_score = random.randint(70, 95)

    else:
        fitness_score = random.randint(60, 90)

    # -------------------------
    # Talent Score Components
    # -------------------------

    batting_impact = (
        runs * 0.002
        + batting_average * 0.30
        + strike_rate * 0.15
    )

    bowling_impact = (
        wickets * 0.08
        + (10 - economy) * 2
    )

    fielding_impact = catches * 0.10

    fitness_impact = fitness_score * 0.25

    raw_score = (
        batting_impact
        + bowling_impact
        + fielding_impact
        + fitness_impact
    )

    data.append([
        player_id,
        age,
        role,
        matches,
        runs,
        batting_average,
        strike_rate,
        fifties,
        hundreds,
        wickets,
        economy,
        catches,
        fitness_score,
        raw_score
    ])

# -------------------------
# Create DataFrame
# -------------------------

columns = [
    "player_id",
    "age",
    "role",
    "matches",
    "runs",
    "batting_average",
    "strike_rate",
    "fifties",
    "hundreds",
    "wickets",
    "economy",
    "catches",
    "fitness_score",
    "raw_score"
]

df = pd.DataFrame(data, columns=columns)

# -------------------------
# Talent Score (0-100)
# -------------------------

scaler = MinMaxScaler()

df["talent_score"] = scaler.fit_transform(
    df[["raw_score"]]
) * 100

df["talent_score"] = df["talent_score"].round(2)

# -------------------------
# Future Potential
# -------------------------

def future_potential(score):

    if score >= 80:
        return "Elite"

    elif score >= 60:
        return "High"

    elif score >= 40:
        return "Medium"

    else:
        return "Low"

df["future_potential"] = df["talent_score"].apply(
    future_potential
)

# -------------------------
# Auction Value (Crores)
# -------------------------

df["auction_value_cr"] = (
    df["talent_score"] * 0.12
    + df["batting_average"] * 0.05
    + df["wickets"] * 0.02
    + np.random.normal(0, 0.5, len(df))
)

df["auction_value_cr"] = (
    df["auction_value_cr"]
    .clip(lower=0.2)
    .round(2)
)

# -------------------------
# Save Dataset
# -------------------------

df.to_csv("players.csv", index=False)

print("Dataset Created Successfully!")
print("\nShape:", df.shape)

print("\nFuture Potential Distribution:")
print(df["future_potential"].value_counts())

print("\nFirst 5 Rows:")
print(df.head())
```
