```python
import streamlit as st
import pandas as pd
import joblib

# -------------------------------
# Page Configuration
# -------------------------------

st.set_page_config(
    page_title="AI Sports Talent Scout",
    page_icon="🏏",
    layout="wide"
)

# -------------------------------
# Load Models
# -------------------------------

clf = joblib.load("models/future_potential_model.pkl")
reg = joblib.load("models/auction_value_model.pkl")

role_encoder = joblib.load("models/role_encoder.pkl")
potential_encoder = joblib.load("models/potential_encoder.pkl")

# -------------------------------
# Header
# -------------------------------

st.title("🏏 AI Sports Talent Scout")

st.markdown("""
Predict a cricket player's future potential and estimated auction value
using Machine Learning.
""")

st.divider()

# -------------------------------
# Input Section
# -------------------------------

col1, col2 = st.columns(2)

with col1:

    st.subheader("👤 Player Information")

    age = st.number_input(
        "Age",
        min_value=18,
        max_value=40,
        value=22
    )

    role = st.selectbox(
        "Role",
        ["All-Rounder", "Batter", "Bowler", "Wicketkeeper"]
    )

    matches = st.number_input(
        "Matches",
        min_value=0,
        max_value=500,
        value=100
    )

    runs = st.number_input(
        "Runs",
        min_value=0,
        value=3000
    )

    batting_average = st.number_input(
        "Batting Average",
        min_value=0.0,
        value=40.0
    )

    strike_rate = st.number_input(
        "Strike Rate",
        min_value=0.0,
        value=130.0
    )

with col2:

    st.subheader("🏏 Performance Metrics")

    fifties = st.number_input(
        "Fifties",
        min_value=0,
        value=10
    )

    hundreds = st.number_input(
        "Hundreds",
        min_value=0,
        value=2
    )

    wickets = st.number_input(
        "Wickets",
        min_value=0,
        value=50
    )

    economy = st.number_input(
        "Economy",
        min_value=0.1,
        value=6.0
    )

    catches = st.number_input(
        "Catches",
        min_value=0,
        value=20
    )

    fitness_score = st.slider(
        "Fitness Score",
        min_value=50,
        max_value=100,
        value=80
    )

# -------------------------------
# Predict Button
# -------------------------------

if st.button("🚀 Predict Talent"):

    # Encode Role

    role_encoded = role_encoder.transform([role])[0]

    # ---------------------------
    # Feature Engineering
    # ---------------------------

    batting_impact = batting_average * strike_rate

    bowling_impact = wickets / economy

    experience_score = matches * age

    consistency_score = fifties + (2 * hundreds)

    # ---------------------------
    # Create Input DataFrame
    # ---------------------------

    input_data = pd.DataFrame([{
        "age": age,
        "role": role_encoded,
        "matches": matches,
        "runs": runs,
        "batting_average": batting_average,
        "strike_rate": strike_rate,
        "fifties": fifties,
        "hundreds": hundreds,
        "wickets": wickets,
        "economy": economy,
        "catches": catches,
        "fitness_score": fitness_score,
        "batting_impact": batting_impact,
        "bowling_impact": bowling_impact,
        "experience_score": experience_score,
        "consistency_score": consistency_score
    }])

    # ---------------------------
    # Predictions
    # ---------------------------

    future_pred = clf.predict(input_data)[0]

    future_potential = (
        potential_encoder.inverse_transform([future_pred])[0]
    )

    auction_value = reg.predict(input_data)[0]

    # ---------------------------
    # Results
    # ---------------------------

    st.divider()

    st.subheader("📊 Prediction Results")

    col3, col4 = st.columns(2)

    with col3:
        st.success(
            f"Future Potential: {future_potential}"
        )

    with col4:
        st.metric(
            label="Predicted Auction Value",
            value=f"₹ {auction_value:.2f} Cr"
        )

    # ---------------------------
    # Scouting Report
    # ---------------------------

    st.subheader("📝 Scouting Report")

    insights = []

    if batting_average > 45:
        insights.append(
            "Strong batting performance."
        )

    if strike_rate > 140:
        insights.append(
            "Aggressive scoring ability."
        )

    if wickets > 150:
        insights.append(
            "Excellent wicket-taking capability."
        )

    if fitness_score > 85:
        insights.append(
            "Outstanding fitness level."
        )

    if catches > 50:
        insights.append(
            "Reliable fielder."
        )

    if not insights:
        insights.append(
            "Balanced player profile."
        )

    for item in insights:
        st.write("✅", item)
```
